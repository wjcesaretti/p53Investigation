## Biological Background

## The Role of p53 in DNA Damage Response

The p53 protein, encoded by the TP53 gene, is a crucial component in the cellular response to DNA damage. It functions as a transcription factor that can trigger a variety of cellular responses [8]:

1. **DNA Repair Activation**: p53 can initiate the transcription of genes involved in DNA repair pathways, facilitating the correction of DNA lesions.
   
2. **Cell Cycle Arrest**: p53 can induce cell cycle arrest, allowing time for DNA repair. If damage is extensive, p53 can prevent cells from entering the synthesis phase of the cell cycle, where DNA replication occurs.

3. **Apoptosis Induction**: In cases of severe DNA damage, p53 can initiate programmed cell death, or apoptosis, to prevent the propagation of damaged DNA.

4. **Senescence Induction**: As an alternative to apoptosis, p53 can also lead cells into a state of permanent arrest known as senescence.

##Rate Equations
Below are the rate equations that govern the proposed mathematical model for Mdm2 activating p53:

$$
\frac{d[\text{p53}]}{dt} = k'_{s53} + k_{s53} \frac{[\text{Mdm2}_{cyt}]^{4}}{J_{s53}^{4} + [\text{Mdm2}_{cyt}]^{4}} - k_{d53} \cdot [\text{p53}]
$$

$$
\frac{d[\text{Mdm2}_{cyt}]}{dt} = k'_{s2} + k_{s2} \frac{[\text{p53}]^{4}}{J_{s2}^{4} + [\text{p53}]^{4}} - k_{i} \cdot [\text{Mdm2}_{cyt}] + k_{o} \cdot [\text{Mdm2}_{nuc}] - k'_{d2} \cdot [\text{Mdm2}_{cyt}]
$$

$$
\frac{d[\text{Mdm2}_{nuc}]}{dt} = k_{i} \cdot [\text{Mdm2}_{cyt}] - k_{o} \cdot [\text{Mdm2}_{nuc}] - k_{d2} \cdot [\text{Mdm2}_{nuc}]
$$

##Understanding the Equations

**The first equation** calculates the rate of change of the concentration of p53 in the system: $\frac{d[\text{p53}]}{dt}$. $k'_{s53}$ is the basal rate of transcription. The basal rate of transcription indicates the concentration of p53 under normal conditions (no DNA damage). The next term, $k_{s53} \frac{[\text{Mdm2}_{cyt}]^{4}}{J_{s53}^{4} + [\text{Mdm2}_{cyt}]^{4}}$, describes how Mdm2 affects the concentration of p53. The first part of the term, $k_{s53}$, is the maximum value of the hill function, $\frac{[\text{Mdm2}_{cyt}]^{4}}{J_{s53}^{4} + [\text{Mdm2}_{cyt}]^{4}}$. The hill function increases the p53 transcription rate when Mdm2 is high and unaffects the transcription rate when Mdm2 is low. Note the Hill coefficient is 4. Lastly, the decay term is just normal decay $k_{d53} \cdot [\text{p53}]$, multiplying the concentration of p53 by a degradation constant.

**The second equation** calculates the rate of change of the concentration of Mdm2 in the cytoplasm: $\frac{d[\text{Mdm2}_{cyt}]}{dt}$. $k'_{s2}$ is the basal rate of transcription of Mdm2 in the cytoplasm. The basal rate indicates the concentration of Mdm2 in a cell under normal conditions (no DNA damage). There is another Hill function (similar to p53) which shows the relationship between Mdm2 and p53: $k_{s2} \frac{[\text{p53}]^{4}}{J_{s2}^{4} + [\text{p53}]^{4}}$. This Hill coefficient is again, 4. Unlike the previous equation, we now have 2 decay terms: $ k_{i} \cdot [\text{Mdm2}_{cyt}]$ and $ k'_{d2} \cdot [\text{Mdm2}_{cyt}]$. Importantly the equation hosts a term $k_{o} \cdot [\text{Mdm2}_{nuc}]$ which describes the import rate of Mdm2 from the nucleus.

This model takes into account two different compartments, the nucleus and cytoplsam of a cell, that Mdm2 is concentrated. Below is a diagram to help visualize the compartmentalization. While the biology is more complex, involving processes like ubiquitination in the nucleus, our model simplifies these details for clarity. Channels in the nucleus permit the transport of molecules like Mdm2 in and out.

![ ](https://drive.google.com/uc?export=view&id=1CtzwBzZKvSp7zodY9tlv-LCOO0BEREbO) From [9]


**The third equation** calculates the rate of change of the concentration of Mdm2 in the nucleus: $\frac{d[\text{Mdm2}_{nuc}]}{dt}$. The first term, $k_{i} \cdot [\text{Mdm2}_{cyt}]$, represents the import rate of Mdm2 from the cytoplasm to the nucleus. The second term, $ k_{o} \cdot [\text{Mdm2}_{nuc}]$, represents the export rate of Mdm2 from the nucleus to the cytoplasm. The final term, $- k_{d2} \cdot [\text{Mdm2}_{nuc}]$, represents the decay rate of Mdm2 in the nucleus.


##DNA Damage Quantification

We also need to quantify DNA damage and understand how it mathematically impacts the system. The model proposes that the change in the amount of DNA damage is described below:

$\frac{d[\text{DNAdamage}]}{dt} = -k_{\text{repair}} \cdot H(\text{DNAdamage})$

$k_{\text{repair}}$ is the repair rate of DNA while $H(\text{DNAdamage})$ signifies whether or not DNA damage is present. It is a Heaviside step function that takes the binary value of 0 or 1 depending on if damage is present or not. Therefore, DNA will only be repaired when it is damaged.

The degradation rate constants from the above equations depend on the presence/absence of DNA damage. The equation $k_{d2} = k'_{d2} \cdot (1 + \text{DNAdamage})$ adjusts the degradtion rate of Mdm2 in the cytoplasm. The constant $k'_{d2}$ is the baseline degradation rate constant, and the term $(1 + \text{DNAdamage})$ suggests that the degradation rate increases linearly with the extent of DNA damage.

$k_{d53} = k'_{d53} + k''_{d53} \cdot G([\text{Mdm2}]^*, \theta, J_1, [\text{p53}]^*, J_2, [\text{p53}]^*)$

This equation adjusts the degradation rate constant of p53, $k_{d53}$, based on the concentration of Mdm2 and p53. The constant $k'{d53}$ is the baseline degradation rate constant for p53, while $k''{d53}$ scales the effect of the function $G$, which is a Goldbeter-Koshland function which helps define the relationship between p53 and Mdm2.

```python
#@title Loading the Model with Constant DNA Damage
#Model adapted from [1]
import tellurium as te
import numpy as np
import matplotlib.pyplot as plt
import csv
from ipywidgets import *
from IPython.display import clear_output
firstmodel = '''

// Created by libAntimony v2.13.2

#all this function does is finds the max of a and b
function MAX(a, b)
  piecewise(a, a >= b, b);
end

function p53_production_0(Ks53, MDM2cyt, n, Js53, Ks54, a,b)
  a + Ks53 + (b*Ks54)*(MDM2cyt^n/(Js53^n + MDM2cyt^n));
end

p53_production_0 is "p53 production"

function mdm2_production(ks2, p53, m, Js2, ks21)
  ks2 + ks21*(p53^m/(Js2^m + p53^m));
end

mdm2_production is "mdm2 production"


#model *Zhang2007___Mechanism_of_DNA_damage_response__Model1()

  // Compartments and Species:
  compartment cyt, nuc;
  species p53 in nuc, MDM2cyt in cyt, MDM2nuc in nuc, $DNAdamage in nuc;

  // Assignment Rules:
  DNAdamage := heav_x*xyz;
  heav_x := piecewise(0, x < 0, piecewise(1, x > 0, 0));
  xyz := MAX(0, Dam0 - krepair*(time - Tdam));
  kd2_0 := kd2*(1 + DNAdamage);
  kd53_0 := kd53 + kd53_*G_u_v_q_r;
  G_u_v_q_r := 2*u*r/((v - u) + v*q + u*r + (((v - u) + v*q + u*r)^2 - 4*u*r*(v - u))^0.5);
  u := MDM2nuc;
  r := J2/p53;
  v := Q;
  q := J1/p53;
  x := time - ModelValue_27;

  // Reactions:
  p53_production:  -> p53; p53_production_0(ks53, MDM2cyt, n, Js53, ks53_,a,b);
  p53_degradation: p53 -> ; nuc*kd53_0*p53;
  MDM2cyt_production:  => MDM2cyt; mdm2_production(ks2, p53, m, Js2, ks2_);
  MDM2cyt_translocation: MDM2cyt -> MDM2nuc; kin*MDM2cyt - kout*MDM2nuc;
  MDM2cyt_degradation: MDM2cyt => ; cyt*kd2_*MDM2cyt;
  MDM2nuc_degradation: MDM2nuc => ; nuc*kd2_0*MDM2nuc;

  // Species initializations:
  p53 = 0.186;
  MDM2cyt = 0.158;
  MDM2nuc = 0.647000000000002;

  // Compartment initializations:
  cyt = 1;
  nuc = 1;

  // Variable initializations:
  a = 0.0
  b = 1.0
  c = 0.0
  ks53 = 0.6;
  Js53 = 0.45;
  kd53 = 0.27;
  n = 4;
  kin = 0.41;
  kout = 0.05;
  kd2 = 0.05;
  ks2 = 0.15;
  Js2 = 0.92;
  m = 4;
  kd53_ = 8.25;
  Q = 0.83;
  J1 = 0.1;
  J2 = 0.1;
  krepair = 0; #0.08
  ks53_ = 2.56;
  ks2_ = 4.23;
  kd2_ = 0.79;
  dd = 2;
  Dam0 = 2;
  Tdam = 0; #!!!!!! 50 actual val
  ModelValue_27 = Tdam;

  // Other declarations:
  #I made all of the values variables, some were constants, as I can now adjust them
  var heav_x, xyz, kd2_0, kd53_0, G_u_v_q_r, u, r, v, q, x;
  var cyt, nuc, ks53, Js53, kd53, n, kin, kout, kd2, ks2, Js2, m, kd53_;
  var Q, J1, J2, krepair, ks53_, ks2_, kd2_, dd, Dam0, Tdam, ModelValue_27;

  // Unit definitions:
  unit length = metre;
  unit area = metre^2;
  unit volume = litre;
  unit time_unit = second;
  unit substance = mole;

  // Display Names:
  time_unit is "time";
  heav_x is "heav(x)";
  ks53 is "ks53'";
  kd53 is "kd53'";
  kd2 is "kd2'";
  kd2_0 is "kd2";
  ks2 is "ks2'";
  kd53_0 is "kd53";
  kd53_ is "kd53''";
  G_u_v_q_r is "G(u,v,q,r)";
  ks53_ is "ks53''";
  ks2_ is "ks2''";
  kd2_ is "kd2''";
  ModelValue_27 is "Initial for Tdam";
  p53_production is "p53 production";
  p53_degradation is "p53 degradation";
  MDM2cyt_production is "MDM2cyt production";
  MDM2cyt_translocation is "MDM2cyt translocation";
  MDM2cyt_degradation is "MDM2cyt degradation";
  MDM2nuc_degradation is "MDM2nuc degradation";
'''
result = te.loada(firstmodel)
result.simulate(0,40,3000,['Time','p53','MDM2cyt','MDM2nuc','DNAdamage'])
result.plot(xtitle='time',ytitle='concentration',title='Time Series')
```

##Understanding the Graph

###Explaining the Shape


In the above graph, we can see two pulses or oscillations. Why do these occur? What is happening in the system to ouput this shape?

DNA damage is introduced at time = 0 and is kept constant throughout the time series. The prescence of DNA damage results in the degradation of the concentration of MDM2 in the nucleus. Since MDM2 levels decrease, the concentration of p53 increases as there less activation of the p53 degradation pathway. This increase in p53, creates a positive feedback loop with MDM2 creation in the cytoplasm, resulting in sharp increases in both p53 and MDM2. As more MDM2 builds up in the cytoplasm, it begins to sift into the nucleus thereby causing the nuclear MDM2 concentration to increase. This increase then brings p53 back to its steady state. Now the MDM2 creation rate also decreases back to its steady state.

###Multiple Pulses

This process occurs for each pulse. But why are there multiple pulses?

DNA damage does not decrease over time, so the presence of constant DNA damage results in the degration of the concentration of MDM2 in the nucleus, resulting in the same pulse again. ## Bifurcation Diagram

A bifurcation diagram plots the steady state values of a variable based on another variable. In our case, we want to find the value of DNA damage that causes pulses in the system. So we can look at the steady state value/s of p53 to measure if our system is oscillating since an osicillating system will have multiple steady state values.

In the time series above we saw that small oscillations began at DNA damage = 1.4. Then when we increased the DNA damage to 1.5, the large oscillations were present. So, the value of DNA damage that creates oscillations is somewhere between 1.4 and 1.5.

A bifurcation plot allows us to zoom into the 1.4-1.5 interval and pinpoint the value at which the oscillations begin. Run the cell below. Can you find the point at which oscillations begin? Anything odd about the graph?

```python
L =[]
result = te.loada(firstmodel)

for w in np.arange(1.3,1.5,.005):
  for i in np.arange(.1,2,.5):
    for j in np.arange(.1,2,.5):
      result.resetAll()
      result.Dam0 = w
      result.p53 = i
      result.MDM2cyt = j
      m = result.simulate(0,100,3000,['p53'])
      m = result.simulate(0,100,3000,['p53'])
      minM = min(m)[0]
      maxM = max(m)[0]


      L.append([w,maxM,minM])

L=np.array(L)
te.plotArray(L,ylabel="p53",xlabel="DNA damage",title='Bifurcation Plot')
```

### Understanding the Graph

While the level of DNA damage is less than 1.375, the system is stable as there is one constant steady state value of p53. However, the single line begins to split into two at 1.400. This is a [Hopf bifurcation](https://en.wikipedia.org/wiki/Hopf_bifurcation) point. A Hopf bifurcation point is the point at which the steady state solution of the system loses stability and oscillations begin [1]. So, the bifurcation plot represents these oscillations with the maximum and minimum values of p53 over each oscillation.

Therefore DNA damage = 1.41 is our bifurcation point. When DNA damage < 1.41, there will not be any oscillations. When DNA damage > 1.41, there will be oscillations. This information will be useful later when we examine the model with DNA decay.

```python
#@title Loading the Model with Decaying DNA damage
#Model adapted from [1]
model = '''

// Created by libAntimony v2.13.2

#all this function does is finds the max of a and b
function MAX(a, b)
  piecewise(a, a >= b, b);
end

function p53_production_0(Ks53, MDM2cyt, n, Js53, Ks54, a,b)
  a + Ks53 + (b*Ks54)*(MDM2cyt^n/(Js53^n + MDM2cyt^n));
end

p53_production_0 is "p53 production"

function mdm2_production(ks2, p53, m, Js2, ks21)
  ks2 + ks21*(p53^m/(Js2^m + p53^m));
end

mdm2_production is "mdm2 production"


#model *Zhang2007___Mechanism_of_DNA_damage_response__Model1()

  // Compartments and Species:
  compartment cyt, nuc;
  species p53 in nuc, MDM2cyt in cyt, MDM2nuc in nuc, $DNAdamage in nuc;

  // Assignment Rules:
  DNAdamage := heav_x*xyz;
  heav_x := piecewise(0, x < 0, piecewise(1, x > 0, 0));
  xyz := MAX(0, Dam0 - krepair*(time - Tdam));
  kd2_0 := kd2*(1 + DNAdamage);
  kd53_0 := kd53 + kd53_*G_u_v_q_r;
  G_u_v_q_r := 2*u*r/((v - u) + v*q + u*r + (((v - u) + v*q + u*r)^2 - 4*u*r*(v - u))^0.5);
  u := MDM2nuc;
  r := J2/p53;
  v := Q;
  q := J1/p53;
  x := time - ModelValue_27;

  // Reactions:
  p53_production:  -> p53; p53_production_0(ks53, MDM2cyt, n, Js53, ks53_,a,b);
  p53_degradation: p53 -> ; nuc*kd53_0*p53;
  MDM2cyt_production:  => MDM2cyt; mdm2_production(ks2, p53, m, Js2, ks2_);
  MDM2cyt_translocation: MDM2cyt -> MDM2nuc; kin*MDM2cyt - kout*MDM2nuc;
  MDM2cyt_degradation: MDM2cyt => ; cyt*kd2_*MDM2cyt;
  MDM2nuc_degradation: MDM2nuc => ; nuc*kd2_0*MDM2nuc;

  // Species initializations:
  p53 = 0.186;
  MDM2cyt = 0.158;
  MDM2nuc = 0.647000000000002;

  // Compartment initializations:
  cyt = 1;
  nuc = 1;

  // Variable initializations:
  a = 0.0
  b = 1.0
  c = 0.0
  ks53 = 0.6;
  Js53 = 0.45;
  kd53 = 0.27;
  n = 4;
  kin = 0.41;
  kout = 0.05;
  kd2 = 0.05;
  ks2 = 0.15;
  Js2 = 0.92;
  m = 4;
  kd53_ = 8.25;
  Q = 0.83;
  J1 = 0.1;
  J2 = 0.1;
  krepair = 0.08;
  ks53_ = 2.56;
  ks2_ = 4.23;
  kd2_ = 0.79;
  dd = 2;
  Dam0 = 2;
  Tdam = 50;
  ModelValue_27 = Tdam;

  // Other declarations:
  #I made all of the values variables, some were constants, as I can now adjust them
  var heav_x, xyz, kd2_0, kd53_0, G_u_v_q_r, u, r, v, q, x;
  var cyt, nuc, ks53, Js53, kd53, n, kin, kout, kd2, ks2, Js2, m, kd53_;
  var Q, J1, J2, krepair, ks53_, ks2_, kd2_, dd, Dam0, Tdam, ModelValue_27;

  // Unit definitions:
  unit length = metre;
  unit area = metre^2;
  unit volume = litre;
  unit time_unit = second;
  unit substance = mole;

  // Display Names:
  time_unit is "time";
  heav_x is "heav(x)";
  ks53 is "ks53'";
  kd53 is "kd53'";
  kd2 is "kd2'";
  kd2_0 is "kd2";
  ks2 is "ks2'";
  kd53_0 is "kd53";
  kd53_ is "kd53''";
  G_u_v_q_r is "G(u,v,q,r)";
  ks53_ is "ks53''";
  ks2_ is "ks2''";
  kd2_ is "kd2''";
  ModelValue_27 is "Initial for Tdam";
  p53_production is "p53 production";
  p53_degradation is "p53 degradation";
  MDM2cyt_production is "MDM2cyt production";
  MDM2cyt_translocation is "MDM2cyt translocation";
  MDM2cyt_degradation is "MDM2cyt degradation";
  MDM2nuc_degradation is "MDM2nuc degradation";


'''
result = te.loada(model)
result.simulate(40, 80, 100,['Time','p53','MDM2cyt','MDM2nuc','DNAdamage'])
result.plot(xtitle='time',ytitle='concentration',title='Time Series')
```

##Understanding the Graph

###Introduction of DNA damage
Instead of DNA damage having a constant value, we will now introduce DNA damage at time = 50 and look for a response from the system. DNA damage will now linearly decay over time. At time = 50, DNA damage is introduced, resulting in a degradation of the concentration of MDM2 in the nucleus.

###System Dynamics Post-Damage
Since MDM2 levels decrease, the concentration of p53 is now allowed to increase as there is a lack of MDM2. This increase in p53, creates a positive feedback loop with MDM2 creation in the cytoplasm, resulting in sharp increases in both p53 and MDM2. As more MDM2 builds up in the cytoplasm, it begins to sift into the nucleus thereby causing the nuclear MDM2 concentration to increase. This increase then brings p53 back to its steady state. Now the MDM2 creation rate also decreases back to its steady state.

The above simulation is with a relatively small amount of DNA damage. We will next explore how different amounts of damage impact the overall structure of the graph.

##Changing the amount of DNA Damage

###Further Examination of Pulses

### Understanding the Pulses

We can now better see the pulses. But what is their purpose? Notice that the peaks of each pulse decreases over time. Why do you think this occurs?

The greater the amount of DNA damage, the more time it takes for the cell to repair. Since the cell cannot repair all of the damage in one pulse, multiple occur until the damage is repaired; each pulse dampens in strength over time since the amount of DNA damage left to repair decreases over time.

The positive/negative feedback loop of MDM2 and p53 causes these pulses to occur.

We can take this system as an example of how biological systems use dynamic mechanisms to respond to environmental stress. This pulsatile behaviour is not uncommon in biology. Other biological examples of this oscillatory response to stress includes our nervous system's response to various stressors like light and temperature and the secretion of hormones[4].




##Changing the p53 production rate
Now we are going to examine the rate of production of p53. Remember that the rate of production of p53 is $$
\frac{d[\text{p53}]}{dt} = k'_{s53} + k_{s53} \frac{[\text{Mdm2}_{cyt}]^{4}}{J_{s53}^{4} + [\text{Mdm2}_{cyt}]^{4}} - k_{d53} \cdot [\text{p53}]
$$

So we are first going to alter the basal rate of p53, $k'_{s53}$. The basal rate is the production rate of p53 when the system is unstimulated.

```python
result = te.loada(model)
def p53Basal(deltaBasal):
    result.reset()
    result.a = deltaBasal
    result.simulate(40, 80, 100,['Time','p53','MDM2cyt','MDM2nuc','DNAdamage'])
    result.plot(xtitle='time',ytitle='concentration',title='Time Series')

# Create the slider and interact function
p53_slider = widgets.FloatSlider(min=0, max=.5, step=0.1, deltaBasal=0)
interact(p53Basal, deltaBasal = p53_slider)
```

We start the slider at deltaBasal = 0, meaning the change in the basal rate from the original model is 0. Then we slowly increase the basal rate. Observe how the graph shifts to the left. This is due to overall increase in p53 concentration, resulting in a faster pulse.

Next we will alter $k_{s53}$, the coefficient of the hill function that linearly influences the amount of p53 produced. By creating a scalar value, **scalar$k_{s53}$**, we can now alter $k_{s53}$ by multiplying it by our scalar. Hence when scalar$k_{s53}$ = 1, then the value of  $k_{s53}$ is the same as the original. But when scalar$k_{s53}$ = 2, then $k_{s53}$ is 2x its original value. Explore this scalar value below.  

```python
result = te.loada(model)
def k53(scalarK53):
    result.reset()
    result.b = scalarK53
    result.simulate(40, 80, 100,['Time','p53','MDM2cyt','MDM2nuc','DNAdamage'])
    result.plot(xtitle='time',ytitle='concentration',title='Time Series')

# Create the slider and interact function
scalarK53slider = widgets.FloatSlider(min=.10, max=1.8, step=.1, scalarK53=1.0)
interact(k53, scalarK53 = scalarK53slider)
```

###Understanding the Changes

Did you notice anything strange? When we decrease the scalar, the system dampens and there is no longer a pulse. When we increase the scalar, the height of the pulse increases. It also condenses. Since we are increasing the change in concentration of p53, the graph should get steeper as we increase the scalar value.

#Phase Plots
Now we will step away from adjusting parameters with sliders and look at phase plots. A phase plot shows all of the possible states of two variables within a system. The plot has arrows, visualizing the paths of the two variables as they change in time.[3]

We look at phase plots to help understand the how/if the system converges to a steady state. In a phase plot, a steady state is when all of the arrows converge to one line or point. Remember from earlier, a system is in a 'steady state' when its variables and rates are constant over time; there is balance.

We will look at the relationships of the 3 main variables in our model: p53, MDM2 in the nucleus and in the cytoplasm. **However with each phase plot, the variable that is not being plotted is set constant at its initial value.** Look out for points or lines of convergence. You can do this by picking a random point on the screen and tracing the direction of the arrow until you cannot anymore. Try it!

```python
#Adapted from 'Toggle Exercise 7.11v2 .ipynb'

#Initialize two matrices of 200x200
Y,X = np.mgrid[0:1:200j, 0:1:200j]
U, V = np.mgrid[0:1:200j, 0:1:200j]
r = te.loada(model)

#Fill the U,V matrix with values of the change in concentrations
for idx in range (200):
    for idy in range (200):
        r.p53 = X[idx,idy];  r.MDM2nuc = Y[idx,idy]
        U[idx,idy] = r["p53'"]
        V[idx,idy] = r["MDM2nuc'"]

#Plot it
plt.subplots(1,2, figsize=(8,6))
plt.subplot(111)
plt.xlabel('p53', fontsize='16')
plt.ylabel('MDM2 nucleus', fontsize='16')
plt.streamplot(X, Y, U, V, density=[2, 2])
plt.plot(.186, .647, '.', color='red', markersize=30)


plt.show()
```

What do you see? Can you find the steady state value? Is there more than one? Notice how all the states converge to around p53 = .2 and MDM2 in the nucleus = .65, this is the fixed point of the plot.  

Now we will look at the phase plot of p53 and MDM2 in the cytoplasm. Compare the plot with the previous one. How are they different? Where are they similar? Are they related? Should they be related?

```python
#Adapted from 'Toggle Exercise 7.11v2 .ipynb'
xmax = 1
ymax = 1
Y,X = np.mgrid[0:xmax:300j, 0:ymax:300j]
U, V = np.mgrid[0:xmax:300j, 0:ymax:300j]
r = te.loada(model)
r.resetAll()
for idx in range (300):
    for idy in range (300):
        r.p53 = X[idx,idy];  r.MDM2cyt = Y[idx,idy]
        U[idx,idy] = r["p53'"]
        V[idx,idy] = r["MDM2cyt'"]

plt.subplots(1,2, figsize=(8,6))
plt.subplot(111)
plt.xlabel('p53', fontsize='16')
plt.ylabel('MDM2 cytoplasm', fontsize='16')
plt.streamplot(X, Y, U, V, density=[2, 2])
plt.plot(.186, .158, '.', color='red', markersize=30)
plt.show()
```

Now we do notice something different about this graph. All of the arrows do not flow to the same point. Notice at around MDM2 = .18 and p53 = .2, there is a fixed point; all of the arrows in the vicinity flow into that point. However, the arrows relatively further away from that point, flow outwards, towards infinity. Why does this occur?

The steady state point at (.2,.18) is our theoretical steady state. But what about the other arrows not pointing towards the steady state? These arrows indicate that the states will not converge to the calculated steady state, but may converge to another steady state value.

Let's zoom out and take a wider look at the phase plot to determine where these arrows are pointing.
